## What is the percentage of people who have a Bachelor's degree?

```python
bachelors_people = (df["education"] == "Bachelors")
true_bachelors = bachelors_people.sum()
percentage_bachelors = (100 * true_bachelors / len(df)).round(1)
```

# What percentage of people with advanced education (`Bachelors`, `Masters`, or `Doctorate`) make more than 50K?

```python
higher_education_people = df["education"].isin(["Bachelors", "Masters", "Doctorate"])
true_higher_education_people = higher_education_people.sum()
higher_education = (100 * true_higher_education_people / len(df)).round(1)
lower_education = 100 - higher_education
```

```python
higher_education_rich_people = df[higher_education_people]
higher_education_rich_people_nb = (higher_education_rich_people["salary"] == ">50K")
higher_education_rich = (100 * higher_education_rich_people_nb.sum() / true_higher_education_people).round(1)
```

## What percentage of people without advanced education make more than 50K?

```python
true_lower_education_nb = lower_education_people = df["education"][~df["education"].isin(["Bachelors", "Masters", "Doctorate"])]
nb_rich_lower = true_lower_education_nb.value_counts().sum()
lower_education_people = df[~df["education"].isin(["Bachelors", "Masters", "Doctorate"])]["salary"] == ">50K"
lower_education_rich_people_nb = lower_education_people.sum()
lower_education_rich = (100 * lower_education_rich_people_nb / nb_rich_lower).round(1)
```

## What is the minimum number of hours a person works per week (hours-per-week feature)?

```python
min_hours_hours = df["hours-per-week"].min()
```

## What percentage of the people who work the minimum number of hours per week have a salary of >50K?

```python
num_min_workers = (df["hours-per-week"] == min_hours_hours).sum()
people_min_hours_fifthy = df["salary"][df["hours-per-week"] == min_hours_hours] == ">50K"
nb_rich_min_hours_per_week = people_min_hours_fifthy.sum()
rich_percentage = (100 * nb_rich_min_hours_per_week / num_min_workers).round(1)
rich_percentage
```

## What country has the highest percentage of people that earn >50K?

```python
people_earning_more_fifthy = df[(df["salary"] == ">50K")]
nb_people_country = df["native-country"].value_counts()
nb_people_fif_country = people_earning_more_fifthy["native-country"].value_counts()
df_country_salary = pd.merge(nb_people_country, nb_people_fif_country, on='native-country', how='outer', suffixes=('_df1', '_df2'))
df_country_salary = df_country_salary.rename(columns={"count_df1": "nb_total", "count_df2":"nb_people_earning_>50K"})
df_country_salary.reset_index(inplace=True)
df_country_salary.sort_values("nb_people_earning_>50K", ascending=False)
# indexi = range(len(df_country_salary))
# df_country_salary.index(indexi)
```

```python
df_country_salary['percentage_country_earning_>50K'] = ((df_country_salary["nb_people_earning_>50K"]/df_country_salary["nb_total"]) * 100).round(1)
df_country_salary.sort_values("percentage_country_earning_>50K", ascending=False)
```
