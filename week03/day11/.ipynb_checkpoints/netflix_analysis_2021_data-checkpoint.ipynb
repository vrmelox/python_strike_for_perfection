## Q.2 En quelle année le plus grand nombre de séries TV et films ont été sortis ? Montrer les graphiques en barres ?

```python
data['Date_N'] = pd.to_datetime(data['Release_Date'], format='%B %d, %Y', errors='coerce')
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['Date_N'].dt.year.value_counts()
```

```python
year = data['Date_N'].dt.year
year_counts = data['Date_N'].dt.year.value_counts()
ax = year_counts.plot(kind='bar', color="red", xlabel="year", ylabel="year_counts", title="L'année avec le plus grand nombre de sorties")
```
